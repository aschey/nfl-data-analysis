```python
files = os.listdir("../scraped-data")
```

```python
def get_city_mascot(team):
    location = [l for l in two_word_locations if team.startswith(l)]
    if len(location) > 0:
        return location[0], " ".join(team.split()[2:])
    return team.split()[0], " ".join(team.split()[1:])


df = None
for cur_file in files:
    week, date, team1, team2 = cur_file.split("_")
    team1 = team1.replace("-", " ").replace("|", "/").replace(".csv", "")
    team2 = team2.replace("-", " ").replace("|", "/").replace(".csv", "")
    team1_current = teams_lookup[team1]
    team2_current = teams_lookup[team2]
    team1_city, team1_mascot = get_city_mascot(team1)
    team2_city, team2_mascot = get_city_mascot(team2)
    team1_current_city, team1_current_mascot = get_city_mascot(team1_current)
    team2_current_city, team2_current_mascot = get_city_mascot(team2_current)
    cur_df = pd.read_csv(f"../scraped-data/{cur_file}")
    cur_df = cur_df.drop(columns=["Unnamed: 0"])
    cur_df = cur_df.rename(
        columns={cur_df.columns[-2]: "team1_score", cur_df.columns[-1]: "team2_score"}
    )
    cur_df["week"] = week.replace("-", " ")
    cur_df["date"] = date.replace("-", " ")
    cur_df["team1_city"] = team1_current_city
    cur_df["team2_city"] = team2_current_city
    cur_df["team1_mascot"] = team1_current_mascot
    cur_df["team2_mascot"] = team2_current_mascot
    cur_df["team1_original_city"] = team1_city
    cur_df["team2_original_city"] = team2_city
    cur_df["team1_original_mascot"] = team1_mascot
    cur_df["team2_original_mascot"] = team2_mascot

    if df is None:
        df = cur_df
    else:
        df = pd.concat([cur_df, df], ignore_index=True)
# df
df.to_csv("../data/scores.csv")
```
