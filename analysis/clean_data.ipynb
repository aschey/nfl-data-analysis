```python
files = os.listdir('../data')
teams = set(teams_lookup[team.replace('.csv', '').replace('-', ' ').replace('|', '/')] for team in chain.from_iterable(file.split('_')[-2:] for file in files))
#teams
```

```python
df = None
for cur_file in files:
    week,date,team1,team2 = cur_file.split('_')
    team1 = team1.replace('-', ' ').replace('|', '/')
    team2 = team2.replace('-', ' ').replace('|', '/')
    cur_df = pd.read_csv(f'data/{cur_file}')
    cur_df = cur_df.drop(columns=['Unnamed: 0'])
    cur_df = cur_df.rename(columns={cur_df.columns[-2]:'team1_score',cur_df.columns[-1]:'team2_score'})
    cur_df['week'] = week.replace('-', ' ')
    cur_df['date'] = date.replace('-', ' ')
    cur_df['team1'] = teams_lookup[team1.replace('-', ' ').replace('|', '/').replace('.csv', '')]
    cur_df['team2'] = teams_lookup[team2.replace('-', ' ').replace('|', '/').replace('.csv', '')]
    if df is None:
        df = cur_df
    else:
        df = pd.concat([cur_df, df], ignore_index = True)
#df
df.to_csv('../data/scores.csv')
```
